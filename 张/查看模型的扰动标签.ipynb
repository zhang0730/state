```python
import torch

# 加载映射文件
pert_map = torch.load("/disk2/cai045/STATE/huggingface/models/ST-Tahoe/pert_onehot_map.pt")

# 查看前10个扰动标签
print("前10个扰动标签:", list(pert_map.keys())[:10])

# 查看总数量
print("扰动标签总数:", len(pert_map))

# 检查特定标签是否存在
print("DMSO_TF是否存在:", "DMSO_TF" in pert_map)
```

```python
list(pert_map.keys())
```

每个键是字符串形式的列表，包含一个元组：(药物名称, 浓度, 单位)

('DMSO_TF', 0.0, 'uM') 代表一个没有添加 DMSO 的实验组，通常用作对照组。 在细胞培养实验中，对照组用于排除溶剂本身对细胞的影响，从而更准确地评估其他实验组中添加的药物或干扰因素的效果。

# 提取药物名称（去重）

```python
import ast

# 提取所有药物名称（不重复）
unique_drugs = set()
for key in pert_map.keys():
    # 将字符串解析为实际列表
    drug_list = ast.literal_eval(key)  # 示例: [('DrugName', 0.05, 'uM')]
    drug_name = drug_list[0][0]        # 获取药物名称
    unique_drugs.add(drug_name)

print("唯一药物数量:", len(unique_drugs))
print("示例药物:", list(unique_drugs)[:5])
```

# 检查特定药物是否存在

```python
# 检查药物是否存在于任何浓度中
target_drug = "Afatinib"
drug_exists = any(ast.literal_eval(key)[0][0] == target_drug for key in pert_map.keys())
print(f"药物 {target_drug} 是否存在:", drug_exists)
```
